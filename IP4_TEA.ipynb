```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Create a simple plot using plt.plot(). Plot a single Python list [3,7,8,9,12].
plotlist=[3,7,8,9,12]
plt.plot(plotlist)
plt.title("Simple Plot of a Single Python List")
plt.show()
```

```python
# 2. create two lists each with 5 numbers in them.
x=[1,2,3,4,5]
y=[10,20,25,30,40]
# 2a) plot them using plt.plot()
plt.plot(x)
plt.plot(y)
plt.title("Question 2: Plot Lists X & Y using plt.plot()")
plt.show()
```

```python
# 2b) create a plot using plt.subplots() and then add x & y on the axes. 
x = [1,2,3,4,5]
y = [10,20,25,30,40] 

fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,4))
fig.suptitle('Question 2b: Plot Lists X & Y using plt.subplots()',fontsize=14)

ax1.plot(x, color='blue')
ax1.set_title('List X')
ax2.plot(y, color='green')
ax2.set_title('List Y')

plt.show()
```

```python
# 3. create a 1D array of 100 evenly spaced numbers between 0 and 100, save it to variable X
X=np.linspace(0,100, num=100)

# 3a. Create a plot using plt.subplots() and plot X versus X2
X_squared=np.square(X)

fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,4))
fig.suptitle('Question 3a: X versus X^2',fontsize=15)
ax1.plot(X, color='blue')
ax1.set_title('X')
ax2.plot(X_squared, color='green')
ax2.set_title('X-Squared')

plt.show()
```

```python
# 3b. Create a scatter plot of X versus the exponential of X (np.exp(X))
expX=np.exp(X)
plt.scatter(X,expX)
plt.title('X versus the Exponential of X')
plt.xlabel('Array X')
plt.ylabel('Exponential of Array X')
plt.show()
```

```python
# 4. Import car-sales.csv  into a DataFrame called car_sales and view.
car_sales=pd.read_csv('car-sales.csv')
car_sales
```

```python
# 4a. Try to plot the 'Price' column using the plot() function
plt.plot(car_sales['Price'])
plt.show()
# This plot has price on the Y-axis and the index of the entry on the X-axis. 
# It doesn't give very much information, since it's not comparing the price to anything.
# We can see that prices range from 4000 to 9700 - that's the most useful information we can glean from this graph.
```

```python
plt.plot(car_sales['Odometer (KM)'])
plt.show()
# This plot has distance traveled from the odometer in KM on the Y-axis and the index of the entry on the X-axis. 
# It doesn't give very much information, since it's not comparing the odometer to anything.
# We can see that the distance on the cars range from about 11,000 KM to over 200,000 KM - that's the most useful information we can glean from this graph.
```

```python
# 5. Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(int64) 
car_sales["Price"]=car_sales["Price"].str.replace("$","")
car_sales["Price"]=car_sales["Price"].str.replace(",","")
car_sales["Price"]=car_sales['Price'].astype('int64')
car_sales
```

```python
# 6. Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column.
TotalSales = np.cumsum(car_sales["Price"])
car_sales['Total Sales']=TotalSales
car_sales
```

```python
# 7. Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today).
car_sales['Sale Date']=pd.date_range(start='2/12/2023', periods=len(car_sales))
car_sales
```

```python
# 8. Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column 
plt.plot(car_sales['Price'],car_sales['Total Sales'])
plt.title('Price vs total sales line plot')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 9. Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.title('Odometer vs. Price scatter plot')
plt.xlabel('Odometer (KM)')
plt.ylabel("Price")
plt.show()
```

```python
# 10. Create a matplotlib bar graph of the 'Make' vs 'Odometer (KM)' columns
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.title("Make vs. Odometer bar plot")
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

```python
# 11. Create a histogram of the 'Make' column
plt.hist(car_sales['Make'])
plt.title("Histogram of car make")
plt.xlabel('Make')
plt.show()
```

```python
# 12. Create a subplot of multiple plots ( step 8,9,10,11).
fig = plt.figure(figsize=(15,10))

fig.suptitle('Plots on Car Data (Questions 8,9,10,11)', 
             fontsize=20)

ax1 = fig.add_subplot(221)
ax1.set_title('Price vs total sales line plot')
ax1.plot(car_sales['Price'],car_sales['Total Sales'])
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')

ax2 = fig.add_subplot(222)
ax2.set_title('Odometer vs. Price scatter plot')
ax2.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel("Price")

ax3 = fig.add_subplot(223)
ax3.set_title("Make vs. Odometer bar plot")
ax3.bar(car_sales['Make'],car_sales['Odometer (KM)'])
ax3.set_xlabel('Make')
ax3.set_ylabel("Odometer (KM)")

ax4 = fig.add_subplot(224)
ax4.set_title("Histogram of car make")
ax4.hist(car_sales['Make'])
ax4.set_xlabel('Make')

plt.show()
```

```python
# 13. Create a box plot of Colour and Total Sales using seaborn plot.
boxplot=sns.boxplot(x='Colour',
                    y='Total Sales',
                    data=car_sales)
```

```python
# 14. Find out the relationship between "Total Sales" and "Odometer (KM)" using linear regression with no error bar and car make will determine the hue
lmplot=sns.lmplot(data=car_sales,
                  x='Total Sales',
                  y='Odometer (KM)',
                  ci=False,
                  hue='Make')
```

```python
# 15. Draw pair plot for all possible combinations.
pplot = sns.pairplot(car_sales)
```

```python
# 16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 
pcor = car_sales[['Total Sales','Price','Odometer (KM)']].corr()
heatmap = sns.heatmap(pcor,annot=True)
```

Correlation comment
- Total sales and price are weakly correlated and it is slightly positive. There is very little relationship between price and total sales.
- Total sales and Odometer are moderately negatively correlated. This means that as total sales decrease, the mileage of the car is somewhat likely to decrease as well. 
- Price and odometer are strongly negatively correlated. this means that as price increases, the distance on the odometer should be decreasing.
